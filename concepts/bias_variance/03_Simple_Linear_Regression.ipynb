```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
# Generate synthetic data: Salary based on years of experience with increasing trend
np.random.seed(42)
num_records = 30
X = np.sort(np.random.randint(0, 51, size=num_records)).reshape(-1, 1)  # Years of experience sorted in ascending order
y_true = np.linspace(800000, 10000000, num=num_records)  # Increasing salary from 800,000 to 10,000,000 linearly
```

```python
# Fit a linear regression model
model = LinearRegression()
model.fit(X, y_true)
y_pred = model.predict(X)
```

```python
# Plot the data and the fitted line
plt.scatter(X, y_true, color='blue', label='True Salary')
plt.plot(X, y_pred, color='red', label='Predicted Salary')
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
plt.title('Linear Regression with Increasing Salary')
plt.legend()
plt.show()
```

```python
# Calculate Mean Absolute Deviation (MAD)
mad = np.mean(np.abs(y_true - y_pred))

# Calculate Mean Squared Deviation (MSD)
msd = np.mean((y_true - y_pred)**2)

# Calculate Root Mean Squared Deviation (RMSD)
rmsd = np.sqrt(msd)

# Print results
print(f'Mean Absolute Deviation (MAD): {mad}')
print(f'Mean Squared Deviation (MSD): {msd}')
print(f'Root Mean Squared Deviation (RMSD): {rmsd}')
```

## Explanation:
Generating Synthetic Data:

X = np.sort(np.random.randint(0, 51, size=num_records)).reshape(-1, 1): Generates num_records (30 in this case) random integers between 0 and 50 inclusive, sorted in ascending order to simulate increasing years of experience.
y_true = np.linspace(800000, 10000000, num=num_records): Generates num_records increasing salary values linearly spaced between 800,000 and 10,000,000.
Model Fitting and Visualization:

We fit a LinearRegression model to predict salaries (y_true) based on years of experience (X).
The model's predictions (y_pred) are plotted against the true salaries (y_true) to visualize how well the model fits the data with the increasing trend.
Error Calculation:

The Mean Absolute Deviation (MAD), Mean Squared Deviation (MSD), and Root Mean Squared Deviation (RMSD) are calculated to evaluate the model's accuracy in predicting salaries based on the simulated data.
This example demonstrates how to generate synthetic data where salary increases with years of experience in a linear fashion. You can adjust the relationship between X and y_true (e.g., using nonlinear relationships, adding noise) to simulate different scenarios and explore the model's behavior under varying conditions.
