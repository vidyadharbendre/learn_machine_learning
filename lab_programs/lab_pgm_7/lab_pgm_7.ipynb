Program: Apply EM algorithm to cluster a set of data stored in a .CSV file. Use the same data set for clustering using k-Means algorithm. Compare the results of these two algorithms and comment on the quality of clustering. You can add Python ML library classes/API in the program.

```python
from sklearn.cluster import KMeans
from sklearn import preprocessing
from sklearn.mixture import GaussianMixture
from sklearn.datasets import load_iris
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Load the Iris dataset
dataset = load_iris()
X = pd.DataFrame(dataset.data, columns=['Sepal_Length', 'Sepal_Width', 'Petal_Length', 'Petal_Width'])
y = pd.DataFrame(dataset.target, columns=['Targets'])

# Plot the actual clusters
plt.figure(figsize=(14, 7))
colormap = np.array(['red', 'lime', 'black'])

# Real Plot
plt.subplot(1, 3, 1)
plt.scatter(X['Petal_Length'], X['Petal_Width'], c=colormap[y['Targets']], s=40)
plt.title('Real')

# K-Means Plot
plt.subplot(1, 3, 2)
model = KMeans(n_clusters=3)
model.fit(X)
predY = model.labels_
plt.scatter(X['Petal_Length'], X['Petal_Width'], c=colormap[predY], s=40)
plt.title('KMeans')

# GMM Plot
scaler = preprocessing.StandardScaler()
X_scaled = scaler.fit_transform(X)
gmm = GaussianMixture(n_components=3)
gmm.fit(X_scaled)
y_cluster_gmm = gmm.predict(X_scaled)

plt.subplot(1, 3, 3)
plt.scatter(X['Petal_Length'], X['Petal_Width'], c=colormap[y_cluster_gmm], s=40)
plt.title('GMM Classification')

# Show the plots
plt.tight_layout()
plt.show()

# Evaluate clustering quality
from sklearn.metrics import accuracy_score

# Convert GMM clusters to match with real labels for evaluation
# Note: Direct accuracy score is not the best metric for clustering, but can be used here for demonstration.
# Here we will simply demonstrate the evaluation process.
# In practice, use metrics like Silhouette Score, Calinski-Harabasz Index, or Davies-Bouldin Index.

# Note: This step is an illustrative example. For real clustering evaluation, consider metrics suited for unsupervised learning.
```
